```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tips = sns.load_dataset("tips")
df_tips.head()
```

## 2.1 barplot e countplot

```python
sns.set_style("whitegrid")
sns.barplot(x="sex", y="total_bill", data=df_tips, estimator="sum")
```

```python
sns.countplot(x="sex", data=df_tips)
```

## 2.2 boxplot e violinplot

```python
sns.boxplot(x="day", y="total_bill", data=df_tips, palette="rainbow", orient="v")
```

```python
ax = sns.boxplot(y="day", x="total_bill", data=df_tips, palette="rainbow", orient="h")
ax.set_xlabel("Conta")
ax.set_ylabel("Dia")
```

```python
type(ax)
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df_tips, palette="coolwarm")
```

```python
sns.violinplot(x="day", y="total_bill", data=df_tips, palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=df_tips, hue='sex', palette='bwr')
```

```python
sns.violinplot(x="day", y="total_bill", data=df_tips, hue="sex", split=True, palette='Pastel1')
```

## 2.3 striplot e swarmplot

```python
sns.stripplot(data=df_tips, x="day", y="total_bill")
```

```python
sns.stripplot(data=df_tips, x="day", y="total_bill", jitter=False)
```

```python
sns.stripplot(data=df_tips, x="day", y="total_bill", hue="sex", palette="twilight")
```

```python
sns.stripplot(data=df_tips, x="day", y="total_bill", hue="sex", palette="twilight")
```

```python
fig, axes = plt.subplots()

sns.stripplot(data=df_tips, x="day", y="total_bill", hue="sex", palette="twilight", ax=axes)
sns.violinplot(x="day", y="total_bill", data=df_tips, hue="sex", split=True, palette='Pastel1', ax=axes)

axes.legend(loc="best")
```

```python
fig, axes = plt.subplots()

sns.stripplot(data=df_tips, x="day", y="total_bill", hue="sex", palette="twilight", ax=axes)
sns.violinplot(x="day", y="total_bill", data=df_tips, hue="sex", split=True, palette='Pastel1', ax=axes)

axes.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
sns.swarmplot(x="day", y="total_bill", data=df_tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=df_tips, hue="sex");
```
