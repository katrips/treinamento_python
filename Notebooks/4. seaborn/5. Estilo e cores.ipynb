# 5. Estilo e cores

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_tips = sns.load_dataset("tips")
```

## 5.1 Estilos

```python
sns.countplot(data=df_tips, x="sex")
```

```python
sns.set_style("white")
sns.countplot(data=df_tips, x="sex")
```

```python
sns.set_style("darkgrid")
sns.countplot(data=df_tips, x="sex")
```

```python
sns.set_style("whitegrid")
sns.countplot(data=df_tips, x="sex")
```

```python
sns.set_style("ticks")
sns.countplot(data=df_tips, x="sex")
```

## 5.2 Remover caixas

```python
sns.countplot(x='sex',data=df_tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=df_tips)
sns.despine(left=True)
```

## 5.3 Tamanho e Aspect Ratio

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=df_tips)
```

```python
sns.lmplot(x='total_bill',y='tip', aspect=4, data=df_tips)
```

## 5.4 Escala e contedo

```python
sns.set_context('poster',font_scale=1)
sns.countplot(x='sex',data=df_tips,palette='coolwarm')
```

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=df_tips,palette='coolwarm')
```

```python
sns.set_context('talk',font_scale=1)
sns.countplot(x='sex',data=df_tips,palette='coolwarm')
```
