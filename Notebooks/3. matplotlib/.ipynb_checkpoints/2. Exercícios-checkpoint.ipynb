```python
path_resultados = "../dados/Copa/results.csv"

df_resultados = pd.read_csv(path_resultados)
df_resultados.head()
```

```python
df_brazil = df_resultados.copy()
df_brazil["date"] = pd.to_datetime(df_brazil["date"], format="%Y-%m-%d")
df_brazil["year"] = df_brazil["date"].dt.year
df_brazil = df_brazil[(df_brazil["home_team"] == "Brazil") | (df_brazil["away_team"] == "Brazil")]

df_brazil.loc[df_brazil[df_brazil["home_team"] == "Brazil"].index, "Gols Brazil"] = df_brazil.loc[df_brazil[df_brazil["home_team"] == "Brazil"].index, "home_score"]
df_brazil["Gols Brazil"] = df_brazil["Gols Brazil"].fillna(df_brazil["away_score"])

df_brazil.head()
```

```python
df_brazil_pt = df_brazil.pivot_table(index="year", values="Gols Brazil", aggfunc="sum", sort="year").reset_index()
df_brazil_pt.head()
```
