# 1. Distribuição

Seaborn é uma biblioteca criada em cima da matplotlib que possuí diversos templates já pré-criados que fornecem uma melhor visão estatística de gráficos

## 1.1 Import

```python
import seaborn as sns
import pandas as pd
```

```python
path_resultados = "../dados/Copa/results.csv"

df_resultados = pd.read_csv(path_resultados)
df_brazil_home = df_resultados[df_resultados["home_team"] == "Brazil"].copy()
df_resultados.head()
```

## 1.2 Data

Seaborn vem com alguns datasets pré-feitos

```python
df_tips = sns.load_dataset("tips")
```

```python
df_tips.head()
```

## 1.3 distplot

Histograma comum

```python
sns.displot(df_tips["total_bill"], bins=20)
```

```python
sns.set_style("whitegrid")
sns.displot(df_tips["total_bill"])
```

```python
sns.displot(df_brazil_home["home_score"], bins = 10)
```

## 1.4 jointplot

jointplot() permite que seja construido um gráfico com 2 gráficos de distribuição

```python
sns.jointplot(
    x="home_score",
    y="away_score",
    data=df_brazil_home,
    kind="scatter" # pode escolher entre scatter, reg, resid, kde e hex
)
```

```python
sns.jointplot(
    x="home_score",
    y="away_score",
    data=df_brazil_home,
    kind="hex" # pode escolher entre scatter, reg, resid, kde e hex
)
```

```python
sns.jointplot(
    x="total_bill",
    y="tip",
    data=df_tips,
    kind="hex" # pode escolher entre scatter, reg, resid, kde e hex
)
```

```python
sns.jointplot(
    x="total_bill",
    y="tip",
    data=df_tips,
    kind="scatter" # pode escolher entre scatter, reg, resid, kde e hex
)
```

```python
sns.jointplot(
    x="total_bill",
    y="tip",
    data=df_tips,
    kind="reg" # pode escolher entre scatter, reg, resid, kde e hex
)
```

## 1.5 pairplot

```python
sns.pairplot(df_tips)
```

```python
sns.pairplot(df_tips, hue="sex", palette="coolwarm")
```

## 1.6 kdeplot

```python
sns.set_style("whitegrid")
sns.displot(df_tips["total_bill"], kde=True)
```

```python
sns.rugplot(df_tips["total_bill"])
```

## 1.7 Regressões

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips, hue="sex")
```

```python
sns.lmplot(
    x='total_bill',
    y='tip',
    data=df_tips,
    hue='sex',
    palette='coolwarm',
    markers=['o','v'],
    scatter_kws={'s':100},
    line_kws={"ls": "--"}
)
```

```python
sns.lmplot(x='total_bill',
           y='tip',
           data=df_tips,
           col='day',
           hue='sex',
           palette='coolwarm',
           aspect=0.6)
```
