# 4. Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_iris = sns.load_dataset('iris')
```

```python
df_iris.head()
```

## 4.1 PairGrid

```python
sns.PairGrid(df_iris)
```

```python
my_grid = sns.PairGrid(df_iris)
my_grid.map(plt.scatter)
```

```python
my_grid = sns.PairGrid(df_iris)
my_grid.map_diag(plt.hist)
my_grid.map_upper(plt.scatter)
my_grid.map_lower(sns.kdeplot)
```

## 4.2 pairplot

```python
sns.pairplot(df_iris)
```

```python
sns.pairplot(df_iris, hue="species")
```

## 5.3 Facet Grid

```python
df_tips = sns.load_dataset("tips")
df_tips.head()
```

```python
grid1 = sns.FacetGrid(
    data=df_tips,
    col="time",
    row="smoker"
)
```

```python
grid1 = sns.FacetGrid(df_tips, col="time", row="smoker")
grid1.map(plt.hist, "total_bill")
```

```python
grid1 = sns.FacetGrid(df_tips, col="time", row="smoker")
grid1.map(plt.scatter, "total_bill", "tip")
```

```python
grid1 = sns.FacetGrid(df_tips, col="time", row="smoker", hue="sex")
grid1.map(plt.scatter, "total_bill", "tip").add_legend()
```

## 5.4 JointGrid

```python
grid1 = sns.JointGrid(x="total_bill", y="tip", data=df_tips)
grid1.plot(sns.regplot, sns.violinplot)
```
