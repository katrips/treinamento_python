# 3 Plots matriciais

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_flights = sns.load_dataset("flights")
df_tips = sns.load_dataset("tips")
df_flights.head()
```

## 3.1 Heatmap - Mapa de calor

```python
df_tips.head()
```

```python
df_tips.corr(numeric_only=True)
```

```python
sns.heatmap(df_tips.corr(numeric_only=True))
```

```python
sns.heatmap(df_tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

```python
df_flights_pt = df_flights.pivot_table(
    index="month",
    columns="year",
    values="passengers",
    aggfunc="sum"
)
df_flights_pt
```

```python
sns.heatmap(df_flights_pt)
```

```python
sns.heatmap(df_flights_pt, cmap='magma', linecolor='white', linewidth=2)
```

## 3.2 clustermap

```python
sns.clustermap(df_flights_pt)
```
